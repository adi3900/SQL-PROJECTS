# Q4. Is the distribution of sugar prices in the US significantly different from a normal distribution?

```python
df['date'] = pd.to_datetime(df['date'])

sugar_us_prices = df['sugar_us']

shapiro_test_stat, shapiro_p_value = stats.shapiro(sugar_us_prices)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(sugar_us_prices, bins=20, edgecolor='k', alpha=0.7)
plt.title('Histogram of US Sugar Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
stats.probplot(sugar_us_prices, dist="norm", plot=plt)
plt.title('Q-Q Plot of US Sugar Prices')

plt.tight_layout()
plt.show()

shapiro_test_stat, shapiro_p_value
```

The distribution of US sugar prices shows significant deviation from normality. This is supported by:

The Shapiro-Wilk test results show a p-value of 0, which is less than 0.05, indicating we reject the null hypothesis that the data is normally distributed

Visual inspection of the plots shows:

The histogram reveals a slightly right-skewed distribution
The Q-Q plot shows deviations from the diagonal line, particularly at the tails
The distribution has:

A slight positive skewness of 0.1032
A positive kurtosis of 0.1095, indicating slightly heavier tails than a normal distribution
In conclusion, US sugar prices do not follow a normal distribution, showing slight right-skewness and heavier tails than would be expected in a normal distribution.
